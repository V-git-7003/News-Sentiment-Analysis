```python
# Picking out P/L for those dates with news articles
# This accounts for news that has in the weekends too!
# News on sat and sun are corelated to P/L on Monday (Works for any other holidays with two days gap too)


P_L_dict = {}
for i in range(len(Vodafone_fin['Date'])):
    if Vodafone_fin['Date'][i].date() in date_list:
        #print(date_list[i])
        P_L_dict[Vodafone_fin['Date'][i].date()] = Vodafone_fin['P/L'][i]
    
for holiday_news in date_list:
    if holiday_news not in Vodafone_fin.values:
        for i in range(len(Vodafone_fin['P/L'])):
            if holiday_news + timedelta(days=1) == Vodafone_fin['Date'][i]:
                P_L_dict[holiday_news] = Vodafone_fin['P/L'][i]
                break
            elif holiday_news + timedelta(days=2) == Vodafone_fin['Date'][i]:
                P_L_dict[holiday_news] = Vodafone_fin['P/L'][i]
                break
        
len(P_L_dict)

for i in sorted(P_L_dict):
    print(i,':',P_L_dict[i])
type(P_L_dict)
len(P_L_dict)
```

```python
# separating positve and negative newsfor training; 
positive_news_dict = {}
negative_news_dict = {}
for i in sorted(P_L_dict):
    
    if P_L_dict[i] >= 0:
        if i in news_dict:
            positive_news_dict[i] = news_dict[i]
         
    else:
        if i in news_dict:
            negative_news_dict[i] = news_dict[i]
          
len(positive_news_dict) 
#len(negative_news_dict)

#pprint.pprint(positive_news_dict)
pprint.pprint(negative_news_dict)
```

```python
# Making bag of positve and negative words for bootstrapping 
Bag_of_neg_words = []
Bag_of_pos_words = []
negative_news_content = negative_news_dict.values()
positive_news_content = positive_news_dict.values()

#print(positive_news_content)

#print(negative_news_content)
for news_on_different_date in negative_news_content:
    #print(x)
    for news_on_same_date in news_on_different_date:
        Bag_of_neg_words.extend(news_on_same_date.split())
    
for news_on_different_date in positive_news_content:
    #print(x)
    for news_on_same_date in news_on_different_date:
        Bag_of_pos_words.extend(news_on_same_date.split())
        

#news = negative_news_dict.values()
#pprint.pprint(news)
#Bag_of_pos_words
```

```python
#labeling postive and negative news and generating artificial news via bootstrapping


gen_news = []
pre_labels = []

for i in range(1000): # 1000 news articles each 200 words are generated.
    if i%2 == 0: # even indices have positive news and odd neg news. 
        gen_news.append(random.sample(Bag_of_pos_words,200))
        pre_labels.append(1)
    else:
        gen_news.append(random.sample(Bag_of_neg_words,200))
        pre_labels.append(0)

gen_shuffle_news = []
labels = []
for i in indices:  #shuffling the news now!
    gen_shuffle_news.append(gen_news[i])
    labels.append(pre_labels[i])
    
    


#pprint.pprint(gen_shuffle_news[0])
pprint.pprint(labels[0:5])       
```

```python
from nltk.corpus import stopwords
import string
stopwords.words('english')[0:10] 
def text_process(mess):
    """
    Takes in a string of text, then performs the following:
    1. Remove all punctuation
    2. Remove all stopwords
    3. Returns a list of the cleaned text
    """
    # Check characters to see if they are in punctuation
    nopunc = [char for char in mess if char not in string.punctuation]
   

    # Join the characters again to form the string.
    nopunc = ''.join(nopunc)
   # print(nopunc)

    # Now just remove any stopwords
    return [word for word in nopunc.split() if word.lower() not in stopwords.words('english')]
```

## TFID
After the counting, the term weighting and normalization can be done with [TF-IDF](http://en.wikipedia.org/wiki/Tf%E2%80%93idf), using scikit-learn's `TfidfTransformer`.

____
### So what is TF-IDF?
TF-IDF stands for *term frequency-inverse document frequency*, and the tf-idf weight is a weight often used in information retrieval and text mining. This weight is a statistical measure used to evaluate how important a word is to a document in a collection or corpus. The importance increases proportionally to the number of times a word appears in the document but is offset by the frequency of the word in the corpus. Variations of the tf-idf weighting scheme are often used by search engines as a central tool in scoring and ranking a document's relevance given a user query.

One of the simplest ranking functions is computed by summing the tf-idf for each query term; many more sophisticated ranking functions are variants of this simple model.

Typically, the tf-idf weight is composed by two terms: the first computes the normalized Term Frequency (TF), aka. the number of times a word appears in a document, divided by the total number of words in that document; the second term is the Inverse Document Frequency (IDF), computed as the logarithm of the number of the documents in the corpus divided by the number of documents where the specific term appears.

**TF: Term Frequency**, which measures how frequently a term occurs in a document. Since every document is different in length, it is possible that a term would appear much more times in long documents than shorter ones. Thus, the term frequency is often divided by the document length (aka. the total number of terms in the document) as a way of normalization: 

*TF(t) = (Number of times term t appears in a document) / (Total number of terms in the document).*

**IDF: Inverse Document Frequency**, which measures how important a term is. While computing TF, all terms are considered equally important. However it is known that certain terms, such as "is", "of", and "that", may appear a lot of times but have little importance. Thus we need to weigh down the frequent terms while scale up the rare ones, by computing the following: 

*IDF(t) = log_e(Total number of documents / Number of documents with term t in it).*

See below for a simple example.

**Example:**

Consider a document containing 100 words wherein the word cat appears 3 times. 

The term frequency (i.e., tf) for cat is then (3 / 100) = 0.03. Now, assume we have 10 million documents and the word cat appears in one thousand of these. Then, the inverse document frequency (i.e., idf) is calculated as log(10,000,000 / 1,000) = 4. Thus, the Tf-idf weight is the product of these quantities: 0.03 * 4 = 0.12.
____
